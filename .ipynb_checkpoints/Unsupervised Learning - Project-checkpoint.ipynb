```python
# Check for outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
```

```python
# Remove any outliers that are z-score < -3 and z-score > 3
from scipy.stats import zscore

# Calculate z-scores
z_scores = zscore(df)

# Define a threshold to identify outliers
threshold = 3

# Remove outliers
df_clean = df[(z_scores < threshold).all(axis=1)]
```

```python
# Apply one-hot encoding on categorical columns
df_clean = pd.get_dummies(df_clean, columns=['Region', 'Channel'])
df_clean.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_clean), columns=df_clean.columns)
```

```python
from sklearn.decomposition import PCA

# Fit PCA on your dataset
pca = PCA().fit(df_scaled)

# Plot the cumulative explained variance
plt.figure(figsize=(10,7))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
plt.grid(True)

# Save the figure
plt.savefig('cumulative_variance.png', bbox_inches='tight', dpi=300)
```

```python
# Selecting components based on the scree plot
n_components = 6
pca = PCA(n_components=n_components)
principalComponents = pca.fit_transform(df_scaled)

# convert to dataframe
principal_df = pd.DataFrame(data = principalComponents, 
                            columns = ['principal component ' + str(i) for i in range(1, n_components+1)])
principal_df.head()
```

```python
# create a dataframe with loadings
feature_names = df_scaled.columns
loadings = pd.DataFrame(pca.components_, columns=feature_names)

# show loadings
loadings
```

# Part II - KMeans Clustering

The objective of the analysis is to group similar products together into clusters based on their attributes such as fresh, milk, grocery, frozen, detergents_paper, and delicatessen. To perform the k-means clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters, initialize the centroids, assign data points to clusters, update the centroids, and repeat until convergence.

```python
# Determine the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(principal_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
# Work with 6 clusters where the graph forms an elbow
k = 3
kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(principal_df)
```

```python
# Visualize the clusters
plt.scatter(principal_df.iloc[:, 0], principal_df.iloc[:, 1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering


silhouette_scores = []  # To store silhouette scores for different number of clusters

# Loop through different values for k (number of clusters)
for k in range(2, 10):  # Check from 2 to 10 clusters
    clusterer = AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage='ward')
    preds = clusterer.fit_predict(principal_df)
    
    score = silhouette_score(principal_df, preds)
    silhouette_scores.append(score)

# Now plot the silhouette_scores against the number of clusters
plt.figure(figsize=(6,6))
plt.plot(range(2,10), silhouette_scores, 'b*-')
plt.xlim([1,10])
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')

# Save the figure
plt.savefig("silhouette_scores.png", dpi=300)

# Plot the figure
plt.show()
```

# Part III - Hierarchical Clustering 

Hierarchical clustering is a popular unsupervised machine learning algorithm that is used to identify patterns and group similar data points together in a hierarchy. The algorithm works by iteratively merging or splitting clusters based on a similarity measure until a dendrogram is formed.

To perform hierarchical clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters using techniques such as dendrogram.

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

best_clusters = 0  # To keep track of the best number of clusters
best_score = -1  # Initial score for comparison

for i in range(2, 10):  # Let's check from 2 to 10 clusters
    # Perform hierarchical clustering
    cluster = AgglomerativeClustering(n_clusters=i, linkage='ward', metric='euclidean')
    cluster_labels = cluster.fit_predict(principal_df)
    
    # Calculate the silhouette score
    score = silhouette_score(principal_df, cluster_labels)
    
    # Print the silhouette score for each number of clusters
    print(f"For n_clusters = {i}, silhouette score is {score}")
    
    # Compare and store the best number of clusters and silhouette score
    if score > best_score:
        best_clusters = i
        best_score = score

print(f"Best number of clusters is {best_clusters} with silhouette score of {best_score}")

# Now, you can perform final clustering with the best number of clusters
cluster_final = AgglomerativeClustering(n_clusters=best_clusters, linkage='ward', metric='euclidean')
final_cluster_labels = cluster_final.fit_predict(principal_df)

# Add the cluster labels to your DataFrame
principal_df['cluster_labels'] = final_cluster_labels

principal_df.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Decide the range of number of clusters you want to try out
range_n_clusters = range(2,10)

silhouette_scores = [] # List to hold silhouette scores

for n_clusters in range_n_clusters:
    # Create an instance of AgglomerativeClustering with n clusters
    clusterer = AgglomerativeClustering(n_clusters=n_clusters)

    # Fit the data
    cluster_labels = clusterer.fit_predict(df_scaled)

    # Calculate the silhouette score and append to the list
    silhouette_avg = silhouette_score(df_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Now plot the silhouette_scores against the number of clusters
plt.figure(figsize=(6,6))
plt.plot(range_n_clusters, silhouette_scores, 'b*-')
plt.xlim([1,10])
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')

# Save the figure
plt.savefig("silhouette_scores_hierarchical.png", dpi=300)

plt.show()
```

```python
# Add the final cluster labels to the cleaned DataFrame
df_clean['cluster_labels'] = final_cluster_labels

# Display the value counts of the cluster labels
print(df_clean['cluster_labels'].value_counts())

# Analyze each cluster
for i in range(best_clusters):
    print(f"Cluster {i}:")
    print(df_clean[df_clean['cluster_labels'] == i].describe())
```

# Part IV - PCA

In this section you are going to perform principal component analysis (PCA) to draw conclusions about the underlying structure of the wholesale customer data. Since using PCA on a dataset calculates the dimensions which best maximize variance, we will find which compound combinations of features best describe customers.

PCA was already carried out. This step is to further analyze the results

```python
def print_top_features(loadings, n_top=5):
    for i, component in loadings.iterrows():
        top_features = component.nlargest(n_top)
        bottom_features = component.nsmallest(n_top)

        print(f"For principal component {i+1}, the top positive features are :")
        print(top_features)
        print()
        
        print(f"For principal component {i+1}, the top negative features are :")
        print(bottom_features)
        print("\n---\n")

# Call the function
print_top_features(loadings)
```
